```python
import pandas as pd

data = {
    'user ID': [1, 2, 3, 4, 5, 6, 7],
    'Movie Name': ['Inception', 'Titanic', 'Inception', 'Avatar', 'Lion king', 'Titanic', 'Avatar'],
    'Rating': [9, 8, 9, 10, 6, 8, 9]
}


df = pd.DataFrame(data)


pivot_table = df.pivot_table(index='Movie Name', values='Rating', aggfunc='mean')

print(pivot_table)
```

```python
s1 = [20,15,10,25,30,35,28,40,45,60]
scores1 = pd.Series(s1)
scores1
```

```python
import matplotlib.pyplot as plt
plt.boxplot(scores1)
```

```python
plt.figure(figsize=(6,2))
plt.title("Boxplot for batsman scores")
plt.xlabel("Scores")
plt.boxplot(scores1,vert=False)
```
