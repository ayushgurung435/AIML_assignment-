#### Fit a linear regression model

```python
import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data=data1).fit()
model1.summary()
```

#### y^=13.8356 + 1.3397 predicted equ
The probability(p-value) for intercept (beta_0) is 0.707>0.05. Therefore the intercept coeff may not be that much significant in prediction , however the p-value for "daily" (beta-1) is 0.00<0.05.Therefore the beta-1 coeff is highly significant and is contributint to prediction

```python
import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data = data1).fit()
model1.summary()
```

```python
model1.params
```

```python
print(f'model t-values:\n{model1.tvalues}\n--------------\nmodel p-values:\n{model1.pvalues}')
```

```python
(model1.rsquared,model1.rsquared_adj)
```

#### Predict on new data point

```python
newdata=pd.Series([200,300,1500])
data_pred=pd.DataFrame(newdata,columns=['daily'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
pred=model1.predict(data1["daily"])
pred
```

```python
data1["Y_hat"]=pred
data1
```

```python
data1["residuals"]=data1["sunday"]-data1["Y_hat"]
data1
```

```python
mse=np.mean((data1['daily']-data1['Y_hat'])**2)
rmse=np.sqrt(mse)
print('MSE: ',mse)
print('RMSE: ',rmse)
```

```python
plt.scatter(data1['Y_hat'],data1['residuals'])
```

#### Observations
- There appears to be no trend and the residuals are randomly placed around the zero error line.
- Hence the assumption of homoscedasticty is satisfied (constant varience in residuals)

```python
import statsmodels.api as sm
sm.qqplot(data1['residuals'],line='45',fit=True)
plt.show()
```
